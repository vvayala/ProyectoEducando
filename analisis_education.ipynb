```python
# Importar librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/kaggle/input/education-analisis/education.csv')

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Gráfico 1
sns.histplot(data['EI'], kde=True, bins=30, color='blue')
plt.title('Distribución del Índice de Educación (EI)')
plt.xlabel('Índice de Educación (EI)')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico 2
sns.lineplot(data=data, x='year', y='EI', ci=None)
plt.title('Evolución del Índice de Educación (Promedio Global)')
plt.xlabel('Año')
plt.ylabel('Índice de Educación')
plt.show()

# Gráfico 3
sns.scatterplot(x='population', y='EI', data=data, alpha=0.7)
plt.title('Relación entre Población y el Índice de Educación')
plt.xlabel('Población')
plt.ylabel('Índice de Educación')
plt.xscale('log')  
plt.show()

# Gráfico 4
niveles = ['college', 'kindergarten', 'school', 'university']
promedios = data[niveles].mean()
sns.barplot(x=promedios.index, y=promedios.values, palette='viridis')
plt.title('Promedio de estudiantes por nivel educativo')
plt.xlabel('Nivel Educativo')
plt.ylabel('Promedio de Estudiantes')
plt.show()

# Gráfico 5
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlación entre Variables')
plt.show()
```
